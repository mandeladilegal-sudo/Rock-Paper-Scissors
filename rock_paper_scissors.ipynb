# Rock, Paper, Scissors

- We import **"Random"** library 
- We ask the user to choose a game
- We loop through the list of choice and if not there print and error message 

### Logic of the game 

* __Rock__ beats Scissors because a rock can break scissors
* __Scissors__ beat paper because scissors can cut paper
* __Paper__ beat Rock because paper can wrap a rock

### How the winner is decided

* **if both players choose the same thing** : it's a tie
* **if one choice naturally defeats the other** : the winner is the one with the strongger choice according to the rules above
* **if neither of the first two conditions applies** : the remaining player loses

```python
# We import the random library 
import random


def play():
    
    # List of games
    options = ['rock', 'paper', 'scissors']


    # We loop through the games till the condition to try again is not "y" then we break out
    while True:

        # Taking input from the user and lower case the character to match thst string we have in our options list
        user_input = input("Choose rock, paper or scissors :").lower()

        # get the right choice by matching the input with the option in the list
        if user_input not in options:
            print("Not a valid choice. try again")
            continue


        # random choice from the computer using the 'random' funtion 
        computer_input = random.choice(options)
        print(f"The computer choice is {computer_input}")

        # Game logic
        if user_input == computer_input:
            print("It's a tie!")
        elif(user_input == "rock" and computer_input == "scissors") or \
            (user_input == "paper" and computer_input == "rock") or \
            (user_input == "scissors" and computer_input == "paper"):
            print("You Win!")
        else:
            print("You lose!")
        again = input("Play again? (y/n): ").lower()
        if again != "y":
            print("Thanks for playing!")
            break

play()

```
